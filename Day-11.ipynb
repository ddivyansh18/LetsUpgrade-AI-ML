# Framing hypothesis :
- H0: There is no significant relation between Attrition and Total working years
- H1 : There is significant relation between Attrition and Total Working Years

```python
#Finding r and p vale
from scipy.stats import pearsonr
atr = np.array(df["Attrition"])
twy = np.array(df["TotalWorkingYears"])
r,p = pearsonr(atr,twy)
print(p)
```

- As p < 0.05, we reject H0 and accept H1 i.e there is significant relation between Attrition and Total Working Years
